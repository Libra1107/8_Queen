```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
BOARD_SIZE = 8
chess_board = [[0 for _ in range(BOARD_SIZE)] for _ in range(BOARD_SIZE)]
solutions = []

def print_solution(board):
    for i in range(8):
        print(" ".join("Q" if board[i][j] else "." for j in range(8)))
    print()

def is_safe(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, 8, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solve_queen(board, col, solutions):
    if col >= 8:
        solutions.append([row[:] for row in board])
        return
    for i in range(8):
        if is_safe(board, i, col):
            board[i][col] = 1
            solve_queen(board, col + 1, solutions)
            board[i][col] = 0

def get_solution():
    solve_queen(chess_board, 0, solutions)
    for sol in solutions:
        print_solution(sol)
```
